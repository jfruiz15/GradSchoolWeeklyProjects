```python
multiline_text = multiline_text.replace('\n', "")
multiline_text
```

```python
clean_text = ""

for char in multiline_text:
    if char == "":
        clean_text += char
    elif char.isalnum():
        clean_text += char
    else:
        clean_text += " "
        
clean_text
```

```python
clean_list = clean_text.split()
clean_list
```

```python
len(clean_list)
```

```python
# finding unique words
unique_words = dict.fromkeys(clean_list)
len(list(unique_words.keys()))
```

```python
for word in clean_list:
    if unique_words[word] is None:
        unique_words[word]= 1
    else:
        unique_words[word] += 1
unique_words
```

```python
top_words = sorted(unique_words.items(), key=lambda key_val_tuple: key_val_tuple[1])
top_words[:25]
```

```python
from itertools import permutations, dropwhile
```

```python
permutations(range(3))
```

```python
for num_tuple in permutations(range(3)):
    print(num_tuple)
    assert isinstance(num_tuple, tuple)
```

```python
for num_tuple in permutations(range(3)):
    print(list(dropwhile(lambda x: x<= 0, num_tuple)))
```

```python
import math
```

```python
def convert_to_number(number_stack):
    number = 0
    for i in range (0, len(number_stack)):
        number += (number_stack.pop() * (math.pow(10, i)))
    return number

for num_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x: x<= 0, num_tuple))
    print(convert_to_number(number_stack))
```

```python
from itertools import zip_longest
```

```python
def dict_from_csv(header, line):
    zipped_line = zip_longest(header, line, fillvalue=None)
    dict= {kv[0]: kv[1] for kv in zipped_line}
    return dict
```

```python
fd = open("Downloads/sales_record.csv", "r")
fd
```

```python
first_line = fd.readline()
header = first_line.replace("\n", "").split(",")
for i, line in enumerate(fd):
    line = line.replace("\n", "").split(",")
    r= dict_from_csv(header, line)
    print(r)
    if i > 10:
        break
```
