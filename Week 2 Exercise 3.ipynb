## Question 1

```python
#Hypersphere Volume: Plot the volume of a unit hypersphere as a function of dimension. Plot for d=1,⋯,50.

import math
import matplotlib.pyplot as plt

x = []
y = []

for dimension in range(1, 51):
    n = float(dimension)
    volume = math.pi**(n/2)/math.gamma(n/2 + 1)
    x.append(n)
    y.append(volume)
    
plt.plot(x, y, marker = 'o', color = 'r')
plt.title('Volume of a unit hypersphere as a function of dimension')
plt.xlabel('d')
plt.ylabel('vol(Sd(1))')

plt.show()
```

```python
# Hypersphere Radius: What value of radius would one need to maintain a hypersphere volume of 1 with increasing d. 
# Plot this value for d=1,⋯,100.

x = []
y = []
dimension = 1

for dimension in range(1, 101):
    n = float(dimension)
    volume = math.pi**(n/2)/math.gamma(n/2 + 1)
    v = float(volume)
    r = (1/v)**(1/n)
    x.append(n)
    y.append(r)

plt.plot(x, y, color = 'b')
plt.title('Volume of a unit hypersphere as a function of radius')
plt.xlabel('d')
plt.ylabel('radius')

plt.show()

    
```

### Fraction of Volume: Assume we have a hypercube of edge length l=2 centered at the origin (0,0,⋯,0). 
### Generate n=10,000 points uniformly at random for increasing dimensionality d=1,⋯,100. 
### Now answer the following questions:

## Part A

### Plot the fraction of points that lie inside the largest hypersphere that can be 
### inscribed inside the hypercube with increasing d. 
### After how many dimensions does the fraction go to essentially zero?

```python
from scipy.spatial.distance import cdist
import scipy
import numpy as np
```

```python
fraction = {}

for d in range (1, 101): 
    l2_points = np.random.uniform(-1,1, size = (10000,d))
    dist = cdist(l2_points, np.expand_dims(np.zeros(d),0), metric = 'euclidean')
    dist_test = np.sum(dist < 1)/10000
    fraction[d] = dist_test
```

```python
disttupe = sorted(fraction.items())

x,y = zip(*disttupe)

plt.plot(x,y)
plt.xlabel("dimension")
plt.ylabel("fraction of points")
plt.title("Fraction of points that lie within the largest hypersphere")
plt.show()
```

## The fraction essentially goes to zero after about 10 dimensions

## Part B
### Plot the fraction of points in the thin shell of width ϵ=0.01 inside the hypercube (i.e., the difference between the outer hypercube and inner hypercube, or the thin shell along the boundaries). 
### What is the trend that you see? After how many dimensions does the fraction of volume in the thin shell go to 100% (use binary search or increase the dimensionality in steps of 10 to answer this. 
### You may use maximum dimensions of up to 2000, and you may use a threshold of 0.0001 to count the volume as essentially being 1 in the shell, 
### i.e., a volume of 0.9999 can be taken to be equal to 1 for finding the smallest dimension at which this happens).


```python
fraction_shell = {}
fraction_shell_inner = {}

for d in range (1, 101):
    points_shell = np.random.uniform(-1, 1, size =(10000,d))
    dist = cdist(points_shell, np.expand_dims(np.zeros(d),0), metric ='euclidean')
    outer_ring = np.sum(dist < 1)
    inner_ring = np.sum(dist < .99)
    
    shell_points = outer_ring - inner_ring
    
    
    fraction_shell[d] = shell_points/10000
    fraction_shell_inner[d]= inner_ring/10000
    
```

```python
def shell_volume(d):
    vol = (1-(.99**d))*10000
    return(vol)
```

```python
fraction_shell = {}

for d in range (1,101):
    points = shell_volume(d)
    fraction = (points/10000)*100
    
    fraction_shell[d]=fraction
```

```python
shelltupe = sorted(fraction_shell.items())

x,y = zip(*shelltupe)

plt.plot(x, y)
plt.xlabel("dimension")
plt.ylabel("fraction of points")
plt.title("Fraction of points that lie within the thin shell")
plt.show()
```

## Diagonals in High Dimensions

```python
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
# generate a sample
sample1 = normal(loc=20, scale=5, size=100000)
sample2 = normal(loc=40, scale=5, size=100000)
sample = hstack((sample1, sample2))
# fit a cdf
ecdf = ECDF(sample)
# get cumulative probability for values
print('P(x<20): %.3f' % ecdf(20))
print('P(x<40): %.3f' % ecdf(40))
print('P(x<60): %.3f' % ecdf(60))
# plot the cdf
pyplot.plot(ecdf.x, ecdf.y)
pyplot.show()
```

```python
# plot the histogram
pyplot.hist(sample, bins=50)
pyplot.show()
```

```python
import numpy as np

diagonal_entries = np.random.randint(low = 0, high = 1, size = 10000)
print(diagonal_entries)
# i*2-1 will map [0,1] -> [2*0-1 == -1, 2*1-1 == 1] == [-1,1]
modified = [1 if i == 1 else -1 for i in diagonal_entries] 
diag = np.diag(modified)
print(diag)
```

```python
sample = hstack((diag))
# fit a cdf
ecdf = ECDF(sample)
# get cumulative probability for values
print('P(x<20): %.3f' % ecdf(20))
print('P(x<40): %.3f' % ecdf(40))
print('P(x<60): %.3f' % ecdf(60))
# plot the cdf
pyplot.plot(ecdf.x, ecdf.y)
pyplot.show()
```
